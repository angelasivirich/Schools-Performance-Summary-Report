```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
schools_total=len(school_data_complete['school_name'].unique())
students_total=len(school_data_complete['student_name'])
budget_total=school_data['budget'].sum()
math_score_avg=school_data_complete['math_score'].mean()
reading_score_avg=school_data_complete['reading_score'].mean()

#create a new data frame with students whoâ€™s math scores are >=70
passing_math_df=school_data_complete[school_data_complete['math_score']>=70]
percentage_passing_math=(len(passing_math_df['math_score'])/students_total)*100

passing_reading_df=school_data_complete[school_data_complete['reading_score']>=70]
percentage_passing_reading=(len(passing_reading_df['reading_score'])/students_total)*100

passing_both_df=school_data_complete[(school_data_complete['reading_score']>=70) & (school_data_complete['math_score']>=70)]
number_passing_both=len(passing_both_df['reading_score'])
percentage_passing_both=(number_passing_both/students_total)*100

district_summary_df=pd.DataFrame({'Total Schools':[schools_total], 'Total Students':students_total,
                                 'Total Budget':budget_total,'Average Math Score':math_score_avg,
                                 'Average Reading Score':reading_score_avg,'% Passing Math':percentage_passing_math,
                                 '% Passing Reading':percentage_passing_reading,'% Overall Passing':percentage_passing_both })

district_summary_df['Total Budget']=district_summary_df['Total Budget'].map('${:.3f}'.format)

district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#create lists for each column that School Summay Dataframe will display
schools_name=list(school_data_complete['school_name'].groupby(school_data_complete['school_name']).unique().str.get(0))
schools_type=list(school_data_complete['type'].groupby(school_data_complete['school_name']).unique().str.get(0))
students_total=list(school_data_complete['student_name'].groupby(school_data_complete['school_name']).count())
school_budget=school_data_complete['budget'].groupby(school_data_complete['school_name']).mean()

#iterate over every value in lists school_budget and school_total, divide the value of school_budget by the value of school_total with same index , and append the result to a new list per_student_budget
per_student_budget=[]
for x in range(len(school_budget)):
    budget=school_budget[x]/students_total[x]
    per_student_budget.append(budget) 

avg_math_score=list(school_data_complete['math_score'].groupby(school_data_complete['school_name']).mean())
avg_reading_score=list(school_data_complete['reading_score'].groupby(school_data_complete['school_name']).mean())

#list of the total number of studets per school with a passing math score
passing_math=list((school_data_complete['math_score']>=70).groupby(school_data_complete['school_name']).sum())

#iterate over every value in lists passing_math and students_total, divide the value of passing_math by the value of students_total with same index , and append the result to a new list percentage_passing_math
percentage_passing_math=[]    
for x in range(len(passing_math)):
    percentage=(passing_math[x]/students_total[x])*100
    percentage_passing_math.append(percentage)
    
#list of the total number of studets per school with a passing reading score
passing_reading=list((school_data_complete['reading_score']>=70).groupby(school_data_complete['school_name']).sum()) 

#iterate over every value in lists passing_reading and students_total, divide the value of passing_reading by the value of students_total with same index , and append the result to a new list percentage_passing_reading    
percentage_passing_reading=[]    
for x in range(len(passing_reading)):
    percentage=(passing_reading[x]/students_total[x])*100
    percentage_passing_reading.append(percentage)
    
#list of the total number of studets per school with a passing math and reading score   
passing_both=list(((school_data_complete['reading_score']>=70)&(school_data_complete['math_score']>=70)).groupby(school_data_complete['school_name']).sum())

#iterate over every value in lists passing_both and students_total, divide the value of passing_both by the value of students_total with same index , and append the result to a new list percentage_passing_both
percentage_passing_both=[]
for x in range(len(passing_both)):
    percentage=(passing_both[x]/students_total[x])*100
    percentage_passing_both.append(percentage)

#ccombine the lists above into a dataframe
school_summary=pd.DataFrame({'School Name':schools_name,'School Type':schools_type,'Total Students':students_total,
                            'Total School Budget':school_budget,'Per Student Budget':per_student_budget, 
                            'Average Math Score':avg_math_score,
                            'Average Reading Score':avg_reading_score,'% Passing Math':percentage_passing_math,
                            '% Passing Reading':percentage_passing_reading, '% Overall Passing':percentage_passing_both})

#Cloning data frame becuase shool summary is going to be used throughout the code
school_summary_formated = school_summary.copy()

#format columns Total School Budget and Per Student Budget, to display dollar sign and two decimal places
school_summary_formated['Total School Budget']=school_summary_formated['Total School Budget'].map('${:.2f}'.format)
school_summary_formated['Per Student Budget']=school_summary_formated['Per Student Budget'].map('${:.2f}'.format)

school_summary_formated=school_summary_formated.reset_index(drop=True).set_index('School Name')
school_summary_formated
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins=[0,584,629,645,675]
group_name=['<$584','$585-629','$630-644','$645-675']

#create a new dataframe with columns from School Summary dataframe
score_spending_df= school_summary[['Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing']].copy()
school_summary['Per Student Budget']=school_summary['Per Student Budget']
#add 'Spending Ranges (Per Student)' column to the data frame and assign a group name to each column based on 'Per Student Budget'
score_spending_df['Spending Ranges (Per Student)']=pd.cut(school_summary['Per Student Budget'],bins, labels=group_name, include_lowest = True)
#group columns based on average Spending Ranges (Per Student)
grouped_score_spending_df= score_spending_df.groupby(['Spending Ranges (Per Student)']).mean()
grouped_score_spending_df
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
bins=[0,1000,2000,5000]
group_name=['Small (<1000)','Medium (1000-2000)','Large (2000-5000)']

#create a new dataframe with columns from School Summary dataframe
score_size_df= school_summary[["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]].copy()
school_summary['Total Students']=school_summary['Total Students'].astype('float')
#add 'School Size' column to the data frame and assign a group name based on 'Per Student Budget'
score_size_df['School Size']=pd.cut(school_summary['Total Students'],bins, labels=group_name, include_lowest = True)
#group columns based on average School Size
grouped_size_spending_df= score_size_df.groupby(['School Size']).mean()
grouped_size_spending_df
```
